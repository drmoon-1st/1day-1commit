# 시스템 프로그래밍 개요
* shell : 리눅스 시스템의 프로그램의 일종, 사용자가 입력한 명령어를 kernel로 전달, 대소문자 구분
* q : 끝내기
* 리눅스 시스템의 경로 명은 "/"로 구분됨
* "max" : 설명 보여줌
* "pwd" : working dir, 현재 디렉토리
* "cd" : 뒤에 경로 명을 적어 이동, "cd ~"는 홈으로 이동
* "." : 현재 디렉토리를 의미
* ".." : 현재 디렉토리의 부모 디렉토리
* "clear", "cls" : 화면 지우기
* "ls(list)" : 현재 디렉토리 안에 있는 모든 파일 출력
  * -l : 파일의 자세한 정보 출력(ls -l)
  * -t : 수정된 시간 별로 정렬
  * -S : 크기로 정렬
  * -h : 파일의 크기를 사람이 읽기 좋게 출력
  * -r : 역순(reverse)
  * 중복으로 사용가능 -> ls -ltr
  * -a : 숨김 파일까지 다 보여줌
  * "*" : 모든 파일(*.c -> c로 끝나는 모든 파일)
* "mkdir" : 디렉토리를 새로 만듬(mkdir name)
* "rmdir" : 디렉토리를 새로 지움(rmdir name)
* "cat" : 대상 파일의 내용을 출력
* "head", "tail" : 처음, 끝 몇 줄만 출력
* "cp" : 파일 카피(ex->cp filename ~/filename2)
* "mv" : 파일 이동 및 이름 변경(ex->mv filename ~/filename2)
* "rm" : 파일 삭제(\rm filename은 질문 없이 삭제. 주의)
