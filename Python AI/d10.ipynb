```python
 import pandas as pd
 import numpy as np
 import matplotlib.pyplot as plt
```

```python
pd_data = pd.read_csv('/content/drive/MyDrive/data.csv', header=0)
pd_data
```

```python
grouped_mean = pd_data.groupby('diagnosis').mean()
grouped_mean = grouped_mean.reset_index()
grouped_mean
```

```python
nf = len(grouped_mean.columns) - 3
f, ax = plt.subplots(6, 5, figsize=(20, 30))


ci = ri = 0
for c in grouped_mean.columns[2:-1]:
  ax[ci][ri].bar(grouped_mean['diagnosis'], grouped_mean[c])
  ax[ci][ri].set_title(c)
  ri += 1
  if ri == 5:
    ri = 0
    ci += 1
```

```python
def norm(x):
  mean = x.mean()
  std = x.std()
  return (x-mean)/std

pd_data_n = pd.DataFrame()
for c in pd_data.columns[2:-1]:
  pd_data_n[c] = norm(pd_data[c])
```
