```python
import torch
```

```python
x = torch.randn(1)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

y = torch.ones_like(x, device=device)

x = x.to(device)

z = x + y

z.to('cpu', torch.double)
```

```python
a = torch.randn(1, 3, 3)
b = torch.randn(1, 3, 3)
print(a)
print(b)
c = torch.cat((a, b), dim=0)
print(c)
```

```python
a = torch.ones(7)
print(a.numpy())
b = a.add(1)
b = a.add_(1)
print(a, b)
```

```python
a = torch.randn(3, 3)
a = a * 3
print(a.requires_grad)
a.requires_grad_(True)
print(a.requires_grad)
b = (a * a).sum()
print(b)
print(b.grad_fn)
```

```python
x = torch.ones(3, 3, requires_grad=True)

y = x + 5
z = y * y
out = z.mean()
print(x, y, z, out)
print("##################")
out.backward()
print(x, y, z, out)
print("##################")
print(x.grad)
```

```python
print(x.requires_grad)
print((x ** 2).requires_grad)

with torch.no_grad():
  print((x ** 2).requires_grad)
```

```python
print(x.requires_grad)
y = x.detach()
print(y.requires_grad)
print(x.eq(y).all())
```
