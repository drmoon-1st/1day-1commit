```python
import json
from pathlib import Path

import numpy as np
import pandas as pd
from scipy import sparse
from tqdm import tqdm
import glob

pd.options.display.width = 180
pd.options.display.max_colwidth = 120

data_dir = Path('/content/drive/MyDrive/kaggle')
```

```python
NUM_TRAIN = 10000

def read_notebook(path):
    return (
        pd.read_json(
            path,
            dtype={'cell_type': 'category', 'source': 'str'})
        .assign(id=path.stem)
        .rename_axis('cell_id')
    )


paths_train = list((data_dir / 'train').glob('*.json'))[:NUM_TRAIN]
notebooks_train = [
    read_notebook(path) for path in tqdm(paths_train, desc='Train NBs')
]
df = (
    pd.concat(notebooks_train)
    .set_index('id', append=True)
    .swaplevel()
    .sort_index(level='id', sort_remaining=False)
)

df
```

```python
nb_id = df.index.unique('id')[6]
print('Notebook:', nb_id)

print('The diisordered notebook:')
nb = df.loc[nb_id, :]
display(nb)
print()
```

```python
df_orders = pd.read_csv('/content/drive/MyDrive/kaggle/train_orders.csv',
                        index_col='id',
                        squeeze=True,).str.split()
df_orders
```

```python
cell_order = df_orders.loc[nb_id]

print('The ordered notebook:')
nb.loc[cell_order, :]
```

```python
def get_ranks(base, derived):
  return [base.index(d) for d in derived]

cell_ranks = get_ranks(cell_order, list(nb.index))
nb.insert(0, 'rank', cell_ranks)

nb
```

```python
from pandas.testing import assert_frame_equal

assert_frame_equal(nb.loc[cell_order, :], nb.sork_values('rank'))
```

```python
df_orders_ = df_orders.to_frame().join(
    df.reset_index('cell_id').groupby('id')['cell_id'].apply(list),
    how='right'
)

ranks = {}
for id_, cell_order, cell_id in df_orders_.itertuples():
  ranks[id_] = {
      'cell_id': cell_id, 'rank': get_ranks(cell_order, cell_id)
  }
  df_ranks = (
      pd.DataFrame
      .from_dict(ranks, orient='index')
      .rename_axis('id')
      .apply(pd.Series.explode)
      .set_index('cell_id', append=True)
  )

  df_ranks
```
