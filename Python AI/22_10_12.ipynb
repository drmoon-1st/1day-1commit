## **k-means**


*   클러스터링 알고리즘
*   각 클러스터의 중심점으로부터의 거리로 데이터를 분류함
*   순서

1.   각 데이터로부터 가장 가까운 중심점을 계산을 통해 결정
2.   각 데이턱 속하는 클러스터로 분류
3.   각 클러스터의 중심을 재계산해서 해당 위치로 클러스터의 중심점을 이동시킴
4.   가장 가까운 클러스터의 중심점을 재계산하고 각 데이터를 클러스터로 재분류
5.   중심점이 수렴할 때까지 반복



*   임의의 위치에 배치되는 중심점에 의해 결과가 달라질 수 있음

## **벨만-포드법**


*   그래프의 최단 경로 문제를 해결하기 위한 알고리즘

*   순서


1.   시작점은 0 나머지 정점은 무한대로 초기 가중치 설정
2.   간선 중 하나를 선택
3.   간선의 한 정점에서 다른 정점으로 이동하기 위한 가중치를 계산(가중치 = 시작점의 가중치 + 간선의 가중치)
4.   계산 결과가 현재 값보다 작으면 가중치를 새 값으로 변경
5.   값이 변경되면 어느 점에서 온 것인지 표시
6.   모든 간선에 반복작업





## **다익스트라 기법**


*   벨만-포드법보다 효율적인 최단경로 탐색기법
*   순서


1.   시작은  0, 나머지는 무한
2.   간선 별 가중치를거쳐 값 계산 후 정점의 값보다 작으면 변경
3.   다른 간선에도 적용

*   왕복의 간선이 존재하여도 탐색가능

## **A*(에이 스타)**

*   다익스트라 기법을 발전시킨 것
*   일반 탐색 순서
1.  미로를 기준으로 모든 간선의 가중치를 1로 설정
2.  다익스트라 기법으로 최단 경로 찾음
*   추정 가중치 설정가능 : 발견적 가중치(고도)
*   A* 순서
1.  시작점에서 갈 수 있는 가중치 계산
2.  가중치 = 도달할 때까지의 가중치 + 발견적 가중치
3.  최소 가중치를 가지는 칸 선택
4.  선택한 칸을 선택완료로 번경
5.  탐색 완료를 한 칸에서 간선의 가중치 계산
6.  반복
*   발견적 가중치가 실제로 소요되는 최단 가중치에 가까울수록 더 효율적으로 미로 탐색 가능
*   가중치가 낮은게 최적, 너무 높으면 탐색 불가
*   게임 프로그래밍에서 플레이어를 쫓는 적의 움직임을 계산할 때에 자주 사용

## **-*dy/dx*-**


*   dx는 분수가 아닌 기호이다
*   델타 x와 dx의 혼동이 올 수 있다
1.  델타 x는 수치로 여겨지며 변화량을 의미한다.
2.  dx는 수치가 아닌 델타 x의 극한을 표현하는 기호이다
3.  lim(델타 y/델타 x) *x -> 0 = dy/dx
*   이 개념을 통해 속력 식을 통한 가속도 등의 유추나 최소/최대값을 구하는 알고리즘에 사용가능
