# 선형회귀

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

torch.manual_seed(1)
```

```python
x_train = torch.tensor([[1], [2], [3]], dtype=torch.float64)
y_train = torch.tensor([[2], [4], [6]], dtype=torch.float64)

W = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

optimizer = optim.SGD([W, b], lr=0.01)

for epoch in range(4001):
  h = W * x_train + b

  cost = torch.mean((h - y_train)**2)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
  if epoch%500 == 0:
    print(W, b)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

torch.manual_seed(1)

x1_train = torch.tensor([[73], [93], [89], [96], [73]], dtype=torch.float64)
x2_train = torch.tensor([[80], [88], [91], [98], [66]], dtype=torch.float64)
x3_train = torch.tensor([[75], [93], [90], [100], [70]], dtype=torch.float64)
y_train = torch.tensor([[152], [185], [180], [196], [142]], dtype=torch.float64)

w1 = torch.zeros(1, requires_grad=True)
w2 = torch.zeros(1, requires_grad=True)
w3 = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

optimizer = optim.SGD([w1, w2, w3, b], lr=1e-5)

num_epochs = 1000
for epoch in range(num_epochs + 1):
  
  h = x1_train * w1 + x2_train * w2 + x3_train * w3 + b

  cost = torch.mean((h - y_train)**2)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epoch%100 == 0:
    print(cost.item())
```

```python
x_train = torch.FloatTensor([[73, 80, 75],
                        [93, 88, 93],
                        [89, 91, 80],
                        [96, 98, 100],
                        [73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])

W = torch.zeros((3, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)

optimizer = optim.SGD([W, b], lr=1e-5)

EPOCHS = 20
for epoch in range(EPOCHS + 1):

  h = x_train.matmul(W) + b

  cost = torch.mean((h - y_train)**2)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  print(f"epoch : {epoch} cost : {cost.item()}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

from torch.utils.data import TensorDataset, DataLoader

x_train = torch.FloatTensor([[73, 80, 75],
                        [93, 88, 93],
                        [89, 91, 80],
                        [96, 98, 100],
                        [73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])

dataset = TensorDataset(x_train, y_train)

dataloader = DataLoader(dataset, batch_size=2, shuffle=True)

model = nn.Linear(3, 1)
optimizer = torch.optim.SGD(model.parameters(), lr=1e-5)

for epoch in range(21):
  for batch_idx, samples in enumerate(dataloader):
    x_train, y_train = samples

    pred = model(x_train)

    cost = F.mse_loss(pred, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    print(epoch, batch_idx+1, cost.item())
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
  def __init__(self):
    self.x_data = [[1], [2], [3]]
    self.y_data = [[2], [4], [6]]

  def __len__(self):
    return len(self.x_data)
  
  def __getitem__(self, idx):
    x = torch.FloatTensor(self.x_data[idx])
    y = torch.FloatTensor(self.y_data[idx])
    return x, y

dataset = CustomDataset()
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)

model = nn.Linear(1, 3)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

EPOCHS = 1000
for epoch in range(EPOCHS + 1):
  for batch_idx, samples in enumerate(dataloader):
    x_train, y_train = samples

    pred = model(x_train)

    cost = F.mse_loss(pred, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch%100 == 0:
      print(epoch, batch_idx, cost.item())
```

# 로지스틱 회귀

```python
import torch
import torchvision.datasets as dsets
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import torch.nn as nn
import matplotlib.pyplot as plt
import random
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

random.seed(777)
torch.manual_seed(777)
if device == "cuda":
  torch.cuda.manual_seed_all(777)

EPOCHS = 15
batch_size = 500

mnist_train = dsets.MNIST(root='mnist/',
                          train=True,
                          transform=transforms.ToTensor(),
                          download=True)
mnist_test = dsets.MNIST(root='mnist/',
                        train=False,
                        transform=transforms.ToTensor(),
                         download=True)

dataloader = DataLoader(dataset=mnist_train,
                        batch_size=batch_size,
                        shuffle=True,
                        drop_last=True)

model = nn.Linear(784, 10, bias=True).to(device)

criterion = nn.CrossEntropyLoss().to(device)
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)

for epoch in range(EPOCHS + 1):

  avg_cost = 0
  total_batch = len(dataloader)

  for X, y in dataloader:
    X = X.view(-1, 28* 28).to(device)
    y = y.to(device)

    optimizer.zero_grad()
    h = model(X)
    cost = criterion(h, y)
    cost.backward()
    optimizer.step()

    avg_cost += cost / total_batch

  print(epoch, avg_cost)

print("finished")
```

```python
with torch.no_grad():
  x_test = mnist_test.test_data.view(-1, 28 * 28).float().to(device)
  y_test = mnist_test.test_labels.to(device)

  pred = model(x_test)
  correct = torch.argmax(pred, 1) == y_test
  acc = correct.float().mean()
  print(acc.item())
```
